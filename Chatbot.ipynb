```python
import openai
```

```python
# API istemcisini başlatalım:
client = openai.OpenAI(api_key = "YOUR-API-KEY")
```

```python
def chat_with_gpt(history_list):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=history_list
    )
    return response.choices[0].message.content.strip()

if __name__ == "__main__":
    history_list = [
        {"role": "system", "content": "You are a helpful assistant."}
    ]

    print("Çıkmak için 'exit' ya da 'q' yazabilirsiniz.\n")

    while True:
        user_input = input("You: ")
        if user_input.lower() in ["exit", "q"]:
            print("Konuşma tamamlandı.")
            break

        history_list.append({"role": "user", "content": user_input})
        response = chat_with_gpt(history_list)
        print("ChatGPT:", response)
        history_list.append({"role": "assistant", "content": response})
```
