```python
import re
import string

# metin ön işleme adımları:
def preprocess(text):
  text = re.sub(r'[^\w\s\']', ' ', text) # noktalama işaretleri (.,!?@# vs.) gibi özel karakterler temizlenir.
  text = re.sub(r' +', ' ', text) # birden fazla boşluğu tek boşluğa indirger
  return text.strip().lower() # Baştaki ve sondaki boşlukları kaldırır, karakterleri küçük harfe çevirir.
```

```python
df["v2_ozellik"] = df["v2_ozellik"].apply(preprocess)
df.head(3)
```

```python
# Feature-target split
X, y = df.drop('v1', axis = 1), df['v1']

# Train-test split (from complete data)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 40)
df_train = pd.concat([X_train, y_train], axis = 1)

# Validation-test split (from test data)
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size = 0.5, random_state = 40)
df_val, df_test = pd.concat([X_val, y_val], axis = 1), pd.concat([X_test, y_test], axis = 1)

len(df_train), len(df_val), len(df_test)
```

```python
# metin ön işleme:

from nltk.tokenize import word_tokenize, RegexpTokenizer
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
import re

# punctuation remove
def remove_punctuation(text):
    punct_str = string.punctuation
    punct_str = punct_str.replace("'", "") # discarding apostrophe from the string to keep the contractions intact
    return text.translate(str.maketrans("", "", punct_str))

# boşlukların kaldırılması
def remove_whitespace(text):
    return text.strip()

# küçük harfe çevirme
def convert_to_lowercase(text):
    return text.lower()

# stopwordsların kaldrılması
stops = stopwords.words("english") # stopwords
# RegexpTokenizer
regexp = RegexpTokenizer("[\w']+")  # Özellikle konuşma dili, kısaltmalar, ve tek tırnaklar içeren metinlerde klasik word_tokenize yerine tercih edilir

def remove_stopwords(text):
    return " ".join([word for word in regexp.tokenize(text) if word not in stops])
def text_normalizer(text):
    text = convert_to_lowercase(text)
    text = remove_whitespace(text)
    text = re.sub('\n' , '', text) # metni tek satıra dönüştürme
    text = re.sub('\[.*?\]', '', text) # köşeli parantezlerin kaldırılması
    text = remove_punctuation(text)
    text = remove_stopwords(text)
    return text

# Implementing text normalization:
df_train_norm, df_val_norm, df_test_norm = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

df_train_norm['v2'] = df_train['v2'].apply(text_normalizer)
df_val_norm['v2'] = df_val['v2'].apply(text_normalizer)
df_test_norm['v2'] = df_test['v2'].apply(text_normalizer)

df_train_norm['v1'] = df_train['v1']
df_val_norm['v1'] = df_val['v1']
df_test_norm['v1'] = df_test['v1']
df_train_norm.head()
df_train_norm.columns = df_train_norm.columns.str.strip()


X_train_norm, y_train = df_train_norm['v2'].tolist(), df_train_norm['v1'].tolist()
X_val_norm, y_val = df_val_norm['v2'].tolist(), df_val_norm['v1'].tolist()
X_test_norm, y_test = df_test_norm['v2'].tolist(), df_test_norm['v1'].tolist()
```

```python
# TF-IDF vectorization
from sklearn.feature_extraction.text import TfidfVectorizer

TfidfVec = TfidfVectorizer(ngram_range = (1, 1))
X_train_tfidf = TfidfVec.fit_transform(X_train_norm)
X_val_tfidf = TfidfVec.transform(X_val_norm)
X_test_tfidf = TfidfVec.transform(X_test_norm)
```
